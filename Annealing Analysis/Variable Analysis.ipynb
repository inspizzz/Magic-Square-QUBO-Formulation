```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
xs = range(3, 10)
binary_bits = []
domain_bits = []
ohe_bits = []

binary_vars = []
domain_vars = []
ohe_vars = []



for square_size in xs:
    POSSIBLE_NUMBERS = [i+1 for i in range(square_size * square_size)]
    bits = math.floor(math.log2(max(POSSIBLE_NUMBERS)) + 1)
    vars = square_size * square_size * bits

    binary_bits.append(bits)
    binary_vars.append(vars)


    bits = square_size * square_size
    vars = square_size * square_size * bits

    ohe_bits.append(bits)
    ohe_vars.append(vars)

    bits = square_size * square_size - 1
    vars = square_size * square_size * bits

    domain_bits.append(bits)
    domain_vars.append(vars)


plt.plot(xs, binary_vars, marker='x', linewidth=1)
plt.plot(xs, ohe_vars, marker='x', linewidth=1)
plt.plot(xs, domain_vars, marker='x', linewidth=1)
plt.legend(['Binary Encoding', 'One Hot Encoding', 'Domain Encoding'])
plt.xlabel('Magic Square Order')
plt.ylabel('Number of Variables ( linear scale )')
plt.title('Quantity of Binary Variables per Encoding Technique vs. Square Size')

plt.hlines(5000, 0, 10, colors='r', linestyles='dashed', linewidth=1)
# plt.yscale('log')
plt.grid(True)
plt.show()
```
