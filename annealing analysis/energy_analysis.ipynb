```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sizes = [2, 3, 4, 5]
energies = []
```

```python
import pandas as pd

for size in sizes:
    data = np.load(f"domain_wall/solutions_{size}.npy", allow_pickle=True)

    energy_array = []
    
    for sample in data.item()["solutions"]:
        energy_array.append(sample["energy"])

    energies.append(energy_array)
```

```python
np.array(energies).shape
```

```python
plt.title("Magic Square Energies per Size Generated")
plt.xlabel("Magic Square Size")
plt.ylabel("Energy")

plt.violinplot(energies, positions=sizes, showmeans=True)
plt.xticks(sizes, sizes)
plt.show()
```
